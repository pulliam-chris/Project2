```python
import numpy as np
import pandas as pd
```

```python
# Read in the separate data files

census_2016 = pd.read_json("cleaned_census\census_2016_pdx.json")
census_2017 = pd.read_json("cleaned_census\census_2017_pdx.json")
census_2018 = pd.read_json("cleaned_census\census_2018_pdx.json")
census_2019 = pd.read_json("cleaned_census\census_2019_pdx.json")
```

```python
frames = [census_2016, census_2017, census_2018, census_2019]
census_merged = pd.concat(frames)
```

```python
census_merged.head(50)
```

```python
census_merged.rename(columns={
    'Median Age': "MedianAge",
    'Household Income': "HouseholdIncome",
    'Per Capita Income': "PerCapitaIncome",
    'Poverty Count': "PovertyCount",
    'Poverty Rate': "PovertyRate"
}, inplace=True)
census_merged
```

```python
# Export file to json
census_merged.to_json("cleaned_census\census_merged.json", orient="records")
```
