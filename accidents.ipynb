```python
# Dependencies and Setup
import pandas as pd
import datetime as dt

# File to Load 
load_accident_data = "data/US_Accidents_Dec20.csv"

# Read from CSV
accidents_df = pd.read_csv(load_accident_data)

accidents_df
```

```python
#accidents_df = accidents_df["Start_Time"].astype('datetime64', copy=True)
accidents_df = accidents_df.astype({'Start_Time': 'datetime64'}) #.dtypes
accidents_df.dtypes
```

```python
most_recent = accidents_df.loc[(accidents_df["Start_Time"] >= pd.to_datetime('2016-01-01')) & (accidents_df["Start_Time"] < pd.to_datetime('2020-01-01'))]
#cutoff = pd.to_datetime('2019-01-01')
#cutoff
#df.sort_values(by=['col1'])
most_recent = most_recent.sort_values(by=['Start_Time'])
most_recent
```

```python
most_recent.to_csv("data\most_recent.csv")
```

```python
portland = most_recent.loc[most_recent["City"] == "Portland"]
portland
```

```python
portland = portland[["Source", "Severity", "Start_Time", "Start_Lat", "Start_Lng", "City", "Zipcode", "Description", "Temperature(F)", "Weather_Condition", "Visibility(mi)", "Wind_Speed(mph)"]]
portland
```

```python
portland_na = portland.dropna()
portland_na
```
