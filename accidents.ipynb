```python
# Dependencies and Setup
import pandas as pd
import datetime as dt

# File to Load 
load_accident_data = "data/US_Accidents_Dec20.csv"

# Read from CSV
accidents_df = pd.read_csv(load_accident_data)

accidents_df
```

```python
#accidents_df = accidents_df["Start_Time"].astype('datetime64', copy=True)
accidents_df = accidents_df.astype({'Start_Time': 'datetime64'}) #.dtypes
accidents_df.dtypes
```

```python
most_recent = accidents_df.loc[(accidents_df["Start_Time"] >= pd.to_datetime('2018-01-01')) & (accidents_df["Start_Time"] < pd.to_datetime('2020-01-01'))]
#cutoff = pd.to_datetime('2019-01-01')
#cutoff
#df.sort_values(by=['col1'])
most_recent = most_recent.sort_values(by=['Start_Time'])
most_recent
```

```python
most_recent.to_csv("data\most_recent.csv")
```

```python
import csv
import json

csvfile = open('data\portland_cleaned.csv', 'r')
jsonfile = open('data\portland_cleaned.json', 'w')

#fieldnames = ("Source", "Severity", "Start_Time", "Start_Lat", "Start_Lng", "City", "Zipcode", "Description", "Temperature(F)", "Weather_Condition", "Visibility(mi)", "Wind_Speed(mph)")
reader = csv.DictReader(csvfile)
for row in reader:
    jsonfile.write('[{accident: {')
    json.dump(row, jsonfile)
    jsonfile.write('}]\n')
```

```python
import csv
import json

csvFilePath = 'data\portland_cleaned.csv'
jsonFilePath = 'data\portland_cleaned.json'

data = {}
with open(csvFilePath) as csvFile:
    csvReader = csv.DictReader(csvFile)
    for rows in csvReader:
        id = rows['ID']
        data[id] = rows

with open(jsonFilePath, 'w') as jsonFile:
    jsonFile.write(json.dumps(data, indent = 4))

```
