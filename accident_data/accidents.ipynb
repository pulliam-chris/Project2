```python
# Dependencies and Setup
import pandas as pd
import datetime as dt

# File to Load 
load_accident_data = "data/US_Accidents_Dec20.csv"

# Read from CSV
accidents_df = pd.read_csv(load_accident_data)

accidents_df
```

```python
accidents_df["Weather_Condition"].unique()
```

```python
accidents_df.dtypes
```

```python
# Caste Start_Time into datetime data type
accidents_df["Start_Time_caste"] = accidents_df["Start_Time"]
accidents_df = accidents_df.astype({'Start_Time_caste': 'datetime64'}) 
#accidents_df = accidents_df['Zipcode'].loc[accidents_df['Zipcode'].str[:5]]
accidents_df.dtypes
```

```python
align_census = accidents_df.loc[(accidents_df["Start_Time_caste"] >= pd.to_datetime('2016-01-01')) & (accidents_df["Start_Time_caste"] < pd.to_datetime('2020-01-01'))]
align_census = align_census.sort_values(by=['Start_Time_caste'])
align_census
```

```python
# Find Portland accidents
portland = align_census.loc[(align_census["City"] == "Portland") & (align_census["State"] == "OR")]
portland
```

```python
# Remove unwanted columns
portland = portland[["Source", "Severity", "Start_Time", "Start_Time_caste", "Start_Lat", "Start_Lng", "City", "State", "Zipcode", "Description", "Temperature(F)", "Weather_Condition", "Visibility(mi)", "Wind_Speed(mph)"]]
portland
```

```python
# Drop rows with nulls in any required fields
portland_na = portland.dropna()
portland_na
```

```python
# Shrink Zipcode to 5 digits if longer
portland_na['Zipcode'] = portland_na['Zipcode'].str[:5]    
portland_na
```

```python
# Rename columns to simplify access in Mongo
portland_na.rename(columns={
    'Temperature(F)': "Temperature",
    'Visibility(mi)': "Visibility",
    'Wind_Speed(mph)': "Wind_Speed"
}, inplace=True)
portland_na
```

```python
# Don't need to run but was part of Marker Cluster map setting
portland_na.to_json("data\portland_cleaned.json", orient="records")
```

```python
# Mongo DB dependencies
import pymongo

# The default port used by MongoDB is 27017
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

# Define the accident database in Mongo
db = client.accidents_db

# Define accidents Collection
accident = db.accidents
```

```python
# Iterature through final Portland dataframe and create Mongo accident collection
for index, row in portland_na.iterrows():
    accident_data={    
        "Source" : row['Source'],
        "Severity" : row['Severity'],
        "Start_Time" : row['Start_Time'],
        "Start_Lat" : row['Start_Lat'],
        "Start_Lng" : row['Start_Lng'],
        "City" : row['City'],
        "Zipcode" : row['Zipcode'],
        "State" : row['State'],
        "Description" : row['Description'],
        "Temperature" : row['Temperature'],
        "Weather_Condition" : row['Weather_Condition'],
        "Visibility" : row['Visibility'],
        "Wind_Speed" : row['Wind_Speed'],
    }
    
    #import into MongoDB         
    accident.insert_one(accident_data)
```

```python
# Retrieve census data from JSON
census_data = pd.read_json("census_data\cleaned_census\census_merged.json")
census_data.dtypes
```

```python
db = client.accidents_db
census = db.census

for index, row in census_data.iterrows():
    zipcode_data={    
        "Zipcode" : row['Zipcode'],
        "Population" : row['Population'],
        "MedianAge" : row['Median Age'],
        "HouseholdIncome" : row['Household Income'],
        "PerCapitaIncome" : row['Per Capita Income'],
        "PovertyCount" : row['Poverty Count'],
        "PovertyRate" : row['Poverty Rate'],
        "Year" : row['Year'],
    }
    
    #import into MongoDB         
    census.insert_one(zipcode_data)
```
